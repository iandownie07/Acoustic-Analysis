```python
from scipy.signal import chirp, spectrogram
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 1500)

w = chirp(t, f0=6, f1=1, t1=10, method='linear')
#  f0 = initial freq, f1 = freq at t1

plt.plot(t, w)

plt.title("Linear Chirp, f(0)=6, f(10)=1")

plt.xlabel('t (sec)')

plt.show()
```
